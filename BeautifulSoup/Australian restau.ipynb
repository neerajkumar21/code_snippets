```python
restaurants = []
links = []
websites = []

for i in range(1, 13):
    source = requests.get('https://www.localbook.com.au/category/restaurants/?page=' + str(i)).text
    soup = BeautifulSoup(source, 'lxml')
    for div in soup.find_all('div', class_ = 'panel panel-default listing_results_result'):
        restaurant = div.h4.text
        link = div.h4.a['href']

        restaurants.append(restaurant)
        links.append(link)
```

```python
links
```

```python
restaurants
```

```python
for link in links:
    try:
        source1 = requests.get(link).text
        soup1 = BeautifulSoup(source1, 'lxml')
        art = soup1.find('div', class_ = 'col-md-6 col-sm-12')
        r = requests.get(art.find_all('a')[1]['href'])
        websites.append(r.url)
    except Exception as e:
        websites.append(None)
```

```python
websites
```

```python
df = pd.DataFrame(list(zip(restaurants, links, websites)), columns = ["Restaurant", "Link", "Website"])
df
```
